### Select by Conditions

Create a column "in_asia" that has the value true or false.

```python
df['in_asia'] = df['continent']=='Asia'
df
```

Select a subset of the df using North America as a filter.

```python
df['continent']=='North America'
```

```python
df[df['continent']=='North America']
```

Select a subset of the df including all countries with population between 100 - 250 million.

```python
mid_pop = df[(df['population']>=1e6) &
            (df['population']<=250000000)]
mid_pop
```

Which countries in Asia have fertility rate smaller than 1.8?

```python
low_fert_asia = df[(df['continent']=='Asia') &
                  (df['fertility']<1.8)]
low_fert_asia.index
```

```python
df[(df['continent']=='Asia') &
                  (df['fertility']<1.8)].index
```

Same as above, but showing fertility and continent.

**df.loc[list_of_rows, list_of_columns]**

```python
low_fert_asia = df.loc[(df['continent']=='Asia') & (df['fertility']<1.8), 
                       ['fertility', 'continent']]
                      
low_fert_asia.head(5)

```
