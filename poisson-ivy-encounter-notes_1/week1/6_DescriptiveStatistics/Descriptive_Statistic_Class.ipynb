# Descriptive Statistics

## Two forms of statistics
Statistics is a set of methods and measures that deal with **collecting**, **cleaning** and **analyzing** a dataset and making quantitative statements about the population. These statements can form two kinds:
<br></br>
+ **Descriptive statistics**: to describe a given data set visually and through metrics. It is a crucial step in every exploratory analysis as we get a better feeling for the data, its characteristics and relationships.
+ **Inferential statistics**: to infer values for attributes of a population by observing a subsample of that population.

## Variable

Characteristic of interest for each person or object in a population* - https://openstax.org/books/introductory-statistics/pages/1-key-terms



### Types of Variables

#### Categorical: Norminal Variables

The data of a nominal variables are sorted into categories and do not show a natural ordering.

+ mode of transportation: (bus, tram, metro...)
+ movie genre (horror, romantic, action)


#### Boolean?

```python
True == 1, False == 0
```

#### Categorical: Ordinal Variables
The values of ordinal variables show a natural ordering (they can but do not need to be numbers). However, the difference between the values is not measurable but allow to class the data in a certain order (better than the other).

+ tall / medium / short
+ ranking in a race (first, second, third)



#### Numerical Variables
The values of metric variables show a natural ordering and the distance between two values is measurable in a certain unit.

+ Distance
+ Cost

### Discrete vs. Continuous Variables


**Discrete/countable number:**  Number of steps in a staircase, number of students.

**Continuous/any values within an interval - not countable:** Time, height, temperature.

### Measures of Central Tendency

### Mean
$$ \bar{x} = \frac{\sum_{i}^{n}x_{i}}{n} \$$



```python
# In pandas
# mean value of Attack feature
df["Attack"].mean()
```

#### Median
The value that divides the sample into two groups, middle number.

The probability of observing a value larger than the median and the probability of observing a value smaller than the median in the sample are 50% each. 

Less prone to outliers than mean because it doesn't care about outliers, just finds the middle value.

Median < Mean - outliers with large values

Median > Mean - outliers with small values

```python
# median value of Attack feature
df["Attack"].median()
```

#### Mode 
it is the most frequent value.

```python
# the most frequent value of Attack feature
df["Attack"].mode()
```

## Histograms

Histograms are used to represent continuous data's frequency with (usually) equal width rectangles. The area of the rectangles are proportional to the frequency of data in each interval (bin).

```python
plt.figure(figsize=(7,4))
df['Attack'].hist() # from pandas 

# from matplotlib.pyplot
plt.plot([df['Attack'].mean()]*2, [1,175], color='green', label='Mean',)
plt.plot([df['Attack'].median()]*2, [0, 175], color='purple', label='Median')
plt.plot([df['Attack'].mode()]*2, [0, 175], color='orange', label='Mode')
plt.legend();
```

### Measures of Dispersion (or Variability)


#### VARIANCE AND STANDARD DEVIATION
These are both measures of variability in the data.

The advantage of the standard deviation is that it's in the same units as the original data and hence is easy to interpret. However, the variance has the advantage that the variances of multiple random variables can be added to find the variance of all of those variables together.

**Standard deviation:** $$ \sigma = \sqrt{\frac{\Sigma(x_i - \mu)^2}{N}} $$

**Variance:** $$ \sigma ^2 $$

Variance is the square of the standard deviation.

```python
df["Attack"].var(), df["Attack"].std(), df["Attack"].mean()
```

### Quantiles 
A quantile defines a particular part of a data set, i.e. a quantile determines the probability of finding values  above or below a certain limit.

One commonly used Quantile is the 4-Quantile or Quartile, which divides the sample into 4 groups of equal size. Two common ways of looking at quantiles in python are q-q plots and boxplots


```python
df["Attack"].quantile() # this assumes 50th percentile, so median.
```

```python
q = df["Attack"].quantile([0.25,0.75]) # 4-Quantile or quartile
q
```

#### Interquartile range



```python
iqr = df['Attack'].quantile(0.75) - df['Attack'].quantile(0.25)
iqr
```

```python
df.describe()
```

```python
sns.boxplot(data=df[['Attack']])
#df[['Attack']].boxplot()
plt.text(x=0.41,y=52,s='Q1',c='red')
plt.text(x=0.41,y=74,s='Q2 or Median',c='purple')
plt.text(x=0.41,y=99,s='Q3',c='red');
plt.text(x=0.41,y=175,s='Outliers');

```

From the documentation:

The box extends from the Q1 to Q3 quartile values of the data,

### Pearson Correlation

It describes the strength of linear relationship between two variables. Between -1 and 1.

```python
fig = plt.figure(figsize = (8,6))
pearson_corr = df.corr(method='pearson')
sns.heatmap(data=pearson_corr,
            xticklabels=pearson_corr.columns,
            yticklabels=pearson_corr.columns,
            cmap='RdBu_r',
            annot=True,
        );
```

### Spearman Correlation

Relationship between variable ranks (better for ordinal variables) as it mainly shows whether they have a monoton relationship (not necessarily linear).

https://statistics.laerd.com/statistical-guides/spearmans-rank-order-correlation-statistical-guide.php

```python
fig = plt.figure(figsize = (8,6))
spearman_corr = df[['Generation','Legendary']].corr(method='spearman')
sns.heatmap(data=spearman_corr, 
            xticklabels=spearman_corr.columns,
            yticklabels=spearman_corr.columns,
            cmap='RdBu_r',
            annot=True);
```

#### Cross tables
Cross tables (also called contingency tables) are frequency tables for two variables. For each possible combination of two variables, they sum up the number of observations. We can then look at the absolute counts or the total, row-wise or column-wise relative frequencies.

```python
cross = pd.crosstab(df['Generation'],df["Legendary"])
cross
```
