## Bash Commands

##### Inspecting directories, files and their contents

`pwd` - prints working directory &rarr; prints the path of the current directory  


`ls` - lists files (and directories) in current directory  
&emsp; `ls -a` - also shows hidden files  
&emsp; `ls -l` - also shows additional information (e.g. `drwxrwxrwx` or `-rw-rw-rw-`)<br>
&emsp;&emsp;**explanation**  
&emsp;&emsp;&emsp;`d` means directory, `-` means that it's a file    
&emsp;&emsp;&emsp;3 times `rwx` means **r**ead, **w**rite, e**x**ecute and refers to `user`, `group`, `others` (3x3 = 9 positions after the `d`)<br>
&emsp;&emsp;&emsp;you can change permissions with `chmod`:<br>
&emsp;&emsp;&emsp;&emsp; e.g. write protect for all `chmod ugo-w <filename>`  
&emsp;&emsp;&emsp;&emsp; e.g. executable for user: `chmod u+x <filename>`  
&emsp; `ls g*` - lists all files starting with `g` (e.g.); `*` denotes any character / any number of characters  
&emsp;&emsp; the dot of the suffix like `.txt`, `.py` is just a part of the file name


`tree` - shows the structure of your directories below the current one.


`cat <filename>` - shows the complete content of a file. (`less <filename>` also works.)   
&emsp; **important remark:** angle bracketed notion, e.g. `<filename>` denotes a placeholder. You must replace it by the literal file name!  


##### Changing, creating directories

`cd <name_of_dir>` - changes directory to <name_of_dir>  
&emsp; `cd ..` - goes one level up (don't miss the space)   
&emsp; `cd ../..` - goes two levels up (don't confuse with `\`)<br>
&emsp; `cd` (without argument) - goes to your home directory, wherever you are currently  


`mkdir <name_of_dir>` - **m**a**k**es (creates) a **dir**ectory <name_of_dir> in the current directory  
&emsp; **remark:** The terms *directory* and *folder* are often used interchangeably.

##### Copying, (re)moving files

`cp file dir/` - copies the file *file* into the directory *dir*  
&emsp; `cp dir/file .` - copies a file in `dir` to current directory (denoted by `.`)<br>
`mv file_name_1 file_name_2` - renames a file  
&emsp; `mv dir1/file dir2/` - moves `file` from directory `dir1` to directory `dir2`.  


`rm file` - deletes a file. ❗❗HEADS UP ❗❗: This cannot be undone. The shell assumes you know what you do!<br>
`rmdir dir_name` - deletes a folder if it is empty.   

##### Helpers

`clear` - visually clears the terminal window from text

&uparrow; - shows previous commands  
`history` - shows the last commands you used  
Tabulator key - completes bash command (e.g. directory names, as far as possible)  
`alias myowncd='cd Documents/spiced_projects'` - alias: Use `myowncd` to actually perform a `cd ` to the given dir  
`Ctrl + c` - interrupts a previously started process or simply cleares the line   
`man <command>` - shows the manual pages of \<command\>. Stop with `q`.
