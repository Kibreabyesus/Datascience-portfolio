```python
def fizz_buzz(x):
    for i in range(1, x+1):
        if i%5==0 and i%3==0:
            print(i,"FizzBuzz")
        elif i%3==0:
            print(i,"Fizz")
        elif i%5==0:
            print(i,"Buzz")

```
