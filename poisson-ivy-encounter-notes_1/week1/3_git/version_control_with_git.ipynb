## Overview

**What is the difference between backup and version control?**

- backup: make sure you don't lose data (e.g. due to hardware failure)
- version control: keep track of changes (e.g. to be able to restore an older version)


**What is git?**  
[git](https://git-scm.com) is one of the world's most popular version control systems. It allows to

- jump back to earlier versions
- edit the same program on multiple computers
- collaborate with other people on a project
- publish your project online
- keep backups of your code

### How does *git* work locally?

0. Do some settings (see below)
1. Decide for / create a directory (`mkdir ...` and `cd ...`) in which your project should live  
    - All files / folders below this level can be tracked.
    - Files / folders above this level will not be part of the project.<br><br>
2. Initialise git at this *working directory* (= create a repository)   
 There are two ways of doing this  
 - `git init` &rarr; initialises git at this empty working directory
 - `git clone ...` &rarr; clones an existing repository to this directory (see below)<br><br>
3. Tell git which files should be added (`git add <file(s)>`) to version control  
 This sets the files to the *staging area* (like a shopping cart of an online shop)<br><br>
4. "commit" these files (`git commit -m "some reminder keywords"`)<br>
 This is the so-called *repository* (or: *repo*) (like: "buy now")
<br><br>
0\. is only done once at all.  
1\. and 2. are only needed once per repo. (You will have 2 repos.)  
while 3. and 4. are usually done frequently.
<br><br>
Optional
- If needed you can restore old versions via `git restore ...`
- You can (and will) create and use so-called branches, e.g. to test things out before deploying   
 `git branch <branch name>` and  
 `git checkout <branch name>`  
 All succeeding `add`s and `commit`s refer to this branch until you `checkout` to another (e.g. master/main) branch.  
 &rarr; You will use this functionality to create your personal branch when you upload (push) your code.



**Why** are staging area and local repo in light green / what is `.git`?

- files / directories starting with a dot are *hidden*. (The are not listed by `ls`, only by `ls -a`)
- All files within `.git` are managed by git and keep the different versions / branches (&rarr; don't touch)

### What about git*hub*? / How does git work remotely?

[github](github.com) is a hosting platform to store the local repo **remotely**

- allows to collaborate with others
- allows to clone repos to other computers
- provides some kind of backup as it is located *in the cloud*


Using github, our above list extends to  
5. Transfer the local repo to github.com via `git push ...`  
6. Or the reverse direction from github.com to the local working copy via `git pull`

## Let's become practical

### First-time / one-off git configuration

**Check installation** 
 `git --version`

**Introduce yourself to git**

- In you terminal / git-bash:

 `git config --global user.name "Your Name"` (`Your Name` can also be a nick name.)

 `git config --global user.email "youremail@yourdomain.com"` 

 `git config --global -l` (if you like to check your credentials)


**Authorization settings**
- generate a **private** (!) and a public key in your ~/.ssh/ folder:

 `ssh-keygen` (You can use an empty passphrase by just hitting the return key.)
 
- copy the **public** key (which has been saved in the file `~/.ssh/id_rsa.pub`) to clipoard
- insert the content of the clipboard to your github profile:
    - login to your github account
    - select Settings (top right) / "SSH und GPG keys" / "New SSH keys"  / give a titel and insert clipboard into "Key" field / hit "Add SSH key"
    
**Clone the two spiced repositories to your PC**

- create a folder (if not done so yesterday already) for data that we (the teachers) provide to you  
 `mkdir ..../spiced_projects/poisson-ivy-encounter-notes`  
 
- open in your browser
 https://github.com/spiced-academy/poisson-ivy-encounter-notes  
 
- copy to clipboard:
 content in: code / ssh (looks like `git@github.com:spiced-academy/poisson-ivy-encounter-notes.git`)
 
- enter in bash at your "poisson-ivy-**encounter-notes**" directory:
 `git clone <clipboard>` to clone the repo
 
- repeat the above steps to clone the "poission-ivy-**student-code**" repository as well to 
 `..../spiced_projects/poisson-ivy-student-code`  
 
**Create a personal branch** 
To be able to discrimitate who of you pushes which code to the **student-code** repo, you need to create a branch.  
(BTW: **No branch creation in the encounter-notes repo**: Everyone of you gets the same files.)  


- on your local PC `cd` to your **student-code** directory and
- `git branch <your name>` to create a branch
- `git checkout <your name>` to switch to that branch (Switching is only needed once, branch is kept even after reboot.)  




### Daily Usage

It is vital that you get into the habit of committing and pushing your work to your branch daily!

 
Make sure you have a **structured directory setup**
- one directory per week
- data folders
- helpers
- ...

**update encounter-notes local repo**

1. `cd` to local repo: `cd poisson-ivy-encounter-notes`
2. download/pull new files: `git pull origin master`  
 &rarr; never `git push` (will not work out)

**copy pulled files from encounter-notes to student-code directory if needed**
Example (you can name your subfolder as you prefer):  
1. `cd` to local repo: `cd poisson-ivy-encounter-notes/week1/<any_subfolder>`
2. inspect the content of this folder: `ls -al`
3. copy newly pulled files. **Adapt this point to your personal preferences**, e.g.:
 - `cp <any file> ../../../my_playground/week1/<any_subfolder>`  
 - work on the files
 - copy them to your studend-code directory, e.g. `cp <finished files> ..../poisson-ivy-student-code/week1/<any_subfolder>`  

hints:  
- `*` is a placeholder for all files and folders (e.g. when using `cp`)<br>
- The `-r` parameter of `cp` (`cp -r`) allows to copy recursively (all files/folders below at once).

**update student-code _local_ repo**

1. go to local repo: `cd poisson-ivy-student-code/<any_subfolder>`
2. check branch: `git branch`
3. switch to your branch if needed (`git checkout <your branch name>`)<br>
 (git remembers the branch, so you don't need to repeat 2. and 3. each time.)
4. stage changes: `git add .` (The dot means all files.)
5. store changes: `git commit -m "<your_message>"`

**update student-code _remote_ repo**

1. upload your changes: `git push origin <your_branch_name>`
2. for the first time: `git push --set-upstream origin <your_branch_name>`   
 &rarr; **NEVER `git pull` here, it may create a big mess!**

### Miscellaneous commands

**show repo status**

- `git status`
- USE OFTEN TO SEE WHAT'S GOING ON

**check available branches**

- local: `git branch`
- local + remote: `git branch -a`

**remove a branch**

- local: `git branch -d <local_branch_name>`
- remote: `git push origin -d <remote_branch_name>`

**show repo history**

`git log`

**file removal**

- pushed already: `git rm <file_name>`
- committed, but not pushed: `git reset --soft` (moves to staging area)
- added only, not committed: `git restore --staged <file_name>`
- not added: remove as usual with GUI/Terminal

**create a repo**
- select your account icon
- 'Your repositories'
- 'New'
- give it a name / select public or private / README; .gitignore; licencse
- clone as seen above
