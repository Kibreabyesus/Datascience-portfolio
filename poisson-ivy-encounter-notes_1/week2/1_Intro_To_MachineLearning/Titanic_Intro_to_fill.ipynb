## Complete Procedure




```python
from sklearn.dummy import DummyClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_wine

```

```python
wine = load_wine()
```

Check what the data looks like.

```python
wine
```

```python
wine_data = wine['data'] # X
wine_target = wine['target'] # y
```

### Split data to train-test

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test  = train_test_split(wine_data ,#X 
                                                     wine_target, #y
                                                     test_size = 0.2,
                                                    random_state = 10)
```

```python
X_train.shape, wine_data.shape, X_test.shape
```

### Initialise the baseline model.

```python
m_bl = DummyClassifier(strategy="most_frequent")
```

#### Train the model

**Fit** the model to the data to get the model parameters.

```python
y_train
```

```python
m_bl.fit(X_train, y_train)
```

#### Make a prediction

Use the input data and the baseline model you just fitted.

```python
m_bl.predict(X_train)
```

#### Calculate a score

**Check the score** for the baseline model. Pass the original data and perfect target values so that the model can compare them to the target values it predicts.

```python
m_bl.score(X_train, y_train)
```

Ok, so it only got it right one third of the time. We will do better in a minute but first let's fix the main thing that we missed.

#### Test-Predict-Score

```python
m_bl.predict(X_test)
m_bl.score(X_test, y_test)
```

For most classification models, the score they return is accuracy.
