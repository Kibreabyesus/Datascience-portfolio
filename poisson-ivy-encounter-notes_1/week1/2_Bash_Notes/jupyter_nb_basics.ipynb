### Kinds of cells / kinds of modes

There are two **types of cells**:<br>
    - code cell like above (execute with `Shift + Return` or `Ctrl + Return`)<br>
    - Markdown cells (for documentation, figures, tables, ...)  


There are **two modes**:<br>
    - blue mode (can be obtained by `Esc`)<br>
    - green mode (can be obtained by `Return` or mouse click into the cell)

### Some commands (in "blue mode")

- dd deletes cell
- a creates cell above
- b creates cell below
- m switches cell to be markdown cell
- y switches cell to be code cell
- h opens help  

How to get help within jupyer notebook:
Just type the command and add a "?"
```python
e.g. df.read_csv?
```
or  hit shift+tab once or twice
